```python
import os
import re
from utils import PreProcess
import pandas as pd
```

```python
directory = 'F:AIS_2023'

#===================Manually modify parameters=======================
cnt = 1
for i in range(1,10):
    folders = [f for f in os.listdir(directory) if os.path.isdir(os.path.join(directory, f)) and f.startswith(f"AIS_2023_0{i}")]

    print(f'==========={i} month, cnt:{cnt}==========')

    # Extract the last two digits of the folder name
    pattern = r'AIS_(\d{4})_(\d{2})_(\d{2})$'
    folder_numbers = [(folder, int(re.match(pattern, folder).group(3))) for folder in folders]

    # Sort by the last two digits of the folder name
    sorted_folders = sorted(folder_numbers, key=lambda x: x[1])

    # Read the CSV files in the sorted folder one by one
    for folder_name, _ in sorted_folders:
        folder_path = os.path.join(directory, folder_name)
        csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
        csv_files.sort()  # Sort by file name
        for csv_file in csv_files:
            csv_file_path = os.path.join(folder_path, csv_file)
            # Process the file here, such as opening the file and reading the content
            print(csv_file_path)
            df = pd.read_csv(csv_file_path, error_bad_lines=False)
            PreProcess(df, cnt)
            cnt+=1
```
